```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
import numpy as np
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
scores = cross_val_score(model, X, y, cv=kf)
```

```python
print("Cross-validation scores:", scores)
```

```python
print("Mean Accuracy:", np.mean(scores))
```

```python
plt.bar(range(1, len(scores)+1), scores, color="skyblue")
plt.axhline(np.mean(scores), color="red", linestyle="--", label=f"Mean = {np.mean(scores):.2f}")
plt.xlabel("Fold")
plt.ylabel("Accuracy")
plt.title("K-Fold Cross Validation Scores")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
y_binary = (iris.target == 0).astype(int)
X = iris.data[:, 0].reshape(-1,1)
log_reg = LogisticRegression()
log_reg.fit(X, y_binary)
X_test = np.linspace(X.min(), X.max(), 300).reshape(-1,1)
y_prob = log_reg.predict_proba(X_test)[:,1]
plt.figure(figsize=(8,6))
plt.scatter(X, y_binary, color="blue", label="Actual Data")
plt.plot(X_test, y_prob, color="red", linewidth=2, label="Logistic Regression Curve")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Probability of Setosa")
plt.title("Logistic Regression on Iris Data")
plt.legend()
plt.grid(True)
plt.show()
```
